```python
#load data 
import pandas as pd 
 
df = pd.read_csv("/content/drive/MyDrive/Data_set/tokyo_2020_tweets.csv")
 
# select only 6000 tweets 
df = df[0:6000]
```

```python
model = BERTopic(verbose=True)
 
#convert to list 
docs = df.text.to_list()
 
topics, probabilities = model.fit_transform(docs)
```
